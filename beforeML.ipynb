# Get the color map

```python
cmap = plt.get_cmap('YlOrRd')
```

# Define our colour indexes u-g and Redshift


```python
x = data['u']-data['g']
r = data['redshift']
```

```python
plt.figure(dpi=300)
plot = plt.scatter(x,r,s = 0.5,lw = 0,c = r,cmap=cmap)
cb = plt.colorbar(plot,label = 'Redshift',orientation = 'vertical')
plt.clim(0,3)
plt.xlabel('Colour Index u-g')
plt.ylabel('Redshift')
```

Note: Very compled correlation between color index vs redshift. 
It requires proper feature selections, Lets select more fetures to find the meaningful correlation.


# Lets use two colour indexes u-g and r-i

```python
x = data['u']-data['g']
y = data['r']-data['i']
r = data['redshift']
```

# Create the plot with plt.scatter and plt.colorbar

```python
plt.figure(dpi=300)
plot = plt.scatter(x,y,s = 0.5,lw = 1,c = r,cmap = cmap)
cb = plt.colorbar(plot,label = 'Redshift',orientation = 'vertical')
plt.clim(0,3)
# Define your axis labels and plot title
plt.xlabel('Colour Index u-g')
plt.ylabel('Colour Index r-i')
plt.legend('Redshift (colour) u-g versus r-i',loc = 1)
plt.title('Redshift (colour) u-g versus r-i')
# Set any axis limits
plt.xlim(-0.5,2.5)
plt.ylim(-0.4,1.0)
plt.savefig('beforeML.png',dpi = 300)

```

Note: We can observe some well defined zone where similar type of glaxy lies , countour map can be used to identify the galaxy for the evaluation of redshift based up on u-g and r-i index...
For better accuracy more index can be added and processed using machine learning as additional dimensions cant be visualize in map....

```python
x = data['u']-data['g']
y = data['g']-data['r']
z = data['r']-data['i']
r = data['redshift']


from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(dpi=250)
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x, y, z, c = r, marker = 'o', s=1)
ax.set_xlabel('u-g')
ax.set_ylabel('g-r')
ax.set_zlabel('r-i')
plt.show()
```
